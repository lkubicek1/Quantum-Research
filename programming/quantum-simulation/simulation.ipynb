```python
from qsim import QuantumCircuit, QuantumRegister
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a quantum register with one qubit
qr = QuantumRegister(1)

# Create a quantum circuit with the quantum register
circuit = QuantumCircuit(qr)
print(qr.get_state())
init_prob_0 = np.abs(qr.get_state()[0, 0]) ** 2
init_prob_1 = np.abs(qr.get_state()[1, 0]) ** 2
print(f"Probability |0⟩: {init_prob_0} | Probability |1⟩: {init_prob_1}")
```

```python
# Visualize the initial state probabilities
plt.bar(['|0⟩', '|1⟩'], [init_prob_0, init_prob_1])
plt.xlabel("Qubit State")
plt.ylabel("Probability")
plt.title("Qubit Initial State Probabilities")
plt.show()
```

```python
# Apply the Pauli-X gate to the qubit
circuit.apply_pauli_x(0)
print(qr.get_state())
prob_0 = np.abs(qr.get_state()[0, 0]) ** 2
prob_1 = np.abs(qr.get_state()[1, 0]) ** 2
print(f"Probability |0⟩: {prob_0} | Probability |1⟩: {prob_1}")
```

```python
# Visualize the state probabilities
plt.bar(['|0⟩', '|1⟩'], [prob_0, prob_1])
plt.xlabel("Qubit State")
plt.ylabel("Probability")
plt.title("Qubit State Probabilities after Pauli-X Gate")
plt.show()
```
